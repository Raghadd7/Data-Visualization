```python
#storing the total SMEs for each year in a pandas DataFrame

visData = pd.DataFrame([[1997,total97],[1998,total98],[1999,total99],[2000,total00],
                      [2001,total01],[2002,total02],
                      [2003,total03],[2004,total04],[2005,total05],[2006,total06],
                      [2007,total07],[2008,total08],[2009,total09],
                      [2010,total10],[2011,total11],[2012,total12],[2013,total13],
                      [2014,total14],[2015,total15],[2016,total16]],
                      columns=['year', 'SMEs'])
```

```python
visData
```

```python
#calculating the percentage change in number of SMEs and storing it in visData

visData['SMEchange']=visData.SMEs.pct_change()*100
```

```python
visData
```

```python
#reading GDP growth rate data for the United states and taking the years of intrest

data = pd.read_csv("API_NY.GDP.MKTP.KD.csv")

data = data.loc[data['Country Name'] == 'United States']

data = data[['1997','1998','1999','2000','2001','2002','2003',
             '2004','2005','2006','2007','2008','2009','2010',
             '2011','2012','2013','2014','2015','2016']]

data=pd.melt(data)
data.columns=['year','GDPgrowth']

#storing GDP growth in visData
visData['GDPgrowth']= data.GDPgrowth 
```

```python
visData
```

```python
# reading Jobs Data and taking only the United States country and the unemployment indicator

data = pd.read_csv("JobsData.csv")

data = data[["Country Name","Indicator Name",
             '1997','1998','1999','2000','2001','2002','2003',
             '2004','2005','2006','2007','2008','2009','2010',
             '2011','2012','2013','2014','2015','2016']]

data = data.loc[data['Country Name'] == 'United States']
data = data.loc[data['Indicator Name'] == 'Unemployment, total (% of total labor force) (modeled ILO estimate)']

del data['Country Name']
del data['Indicator Name']

data=pd.melt(data)

data.columns=['year','Unemployment']

#storing unemployment in visData
visData['Unemployment']= data.Unemployment
```

```python
visData
```

```python
#removing first row with NaN value
visData=visData.iloc[1:]
```

```python
x = visData.year

layout = go.Layout(
    title= ' U.S. Small and Medium-sized Enterprises growth in relation to economic indicators ',
    yaxis=dict(
        title='Annual Percentage Change',
        tickmode='linear'
    ),
   
    xaxis=dict(
        title='year',
        tickmode='linear'
    ),
    
    updatemenus=list([
        dict(
            y=1.07,
            buttons=list([   
                dict(
                    args=[{'visible': [True,True,True]}],
                    label='Both indicators',
                    method='update'
                ),
                dict(
                    args=[{'visible': [False,True, True]}],
                    label='Gross Domestic<br>Product',
                    method='update'
                ),
                dict(
                    args=[{'visible': [True,False,True]}],
                    label='Unemployment',
                    method='update'
                )            
            ])
        ) 
])
    
)


trace1 = go.Scatter(
        x=x,
        y=visData.SMEchange,    
        name='SME growth',
        mode = 'lines',
     marker=dict(
            color= '#1f77b4'
            ),
    opacity = 1
        )

trace2 = go.Scatter(
        x=x,
        y=visData.GDPgrowth,
        name='GDP growth',
        mode = 'lines',
     marker=dict(
            color= '#d62728'
            ),
    
    opacity= 0.7

)



trace3 = go.Scatter(
        x=x,
        y=visData.Unemployment,
        name='Unemployment rate',
        mode = 'lines',
         marker=dict(
            color= '#f9812a'
             #e07117'
             #ff7f0e'
            ),
    opacity=1
        )


       
data = [trace3,trace2,trace1]
   
fig = go.Figure(data=data,layout=layout)

py.offline.iplot(fig)

```
